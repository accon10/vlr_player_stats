```python
import pandas as pd

'''Have multiple csv files from each region, need to merge them together'''

'''
1=europe
2=north america
3=brazil
4=asia pacific
5=korea
6=china
7=japan
8=latin america-south
9=latin america-north
10=oceania
11=mena
12=gc
'''

df1  = pd.read_csv('player_data_eu.csv')
df2  = pd.read_csv('player_data_na.csv')
df3  = pd.read_csv('player_data_br.csv')
df4  = pd.read_csv('player_data_apa.csv')
df5  = pd.read_csv('player_data_kor.csv')
df6  = pd.read_csv('player_data_chn.csv')
df7  = pd.read_csv('player_data_jpn.csv')
df8  = pd.read_csv('player_data_las.csv')
df9  = pd.read_csv('player_data_lan.csv')
df10 = pd.read_csv('player_data_oce.csv')
df11 = pd.read_csv('player_data_mena.csv')

df_list = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11]
```

```python
# merging all the data frames together:

df = pd.concat(df_list, ignore_index=True)
```

```python
df.shape[0]
```

```python
# some rows have missing entries, removing those...

df = df.dropna()
df.shape[0]
```

```python
# parsing the "Matches" column

df['matches_percent'] = df['Matches'].apply(lambda x: int(x.split(' ')[1].replace('%', '')))
df['match_number'] = df['Matches'].apply(lambda x: int(x.split(' ')[0].replace('(', '').replace(')', '')))
df['KAST_percent'] = df['KAST'].apply(lambda x: int(x.replace('%', '')))
# Need to convert all of the strings to actual numbers...
df['Rounds'] = df['Rounds'].apply(lambda x: int(x))
df['Rating'] = df['Rating'].apply(lambda x: float(x))
df['ACS']    = df['ACS'].apply(lambda x: float(x))
df['KD']     = df['KD'].apply(lambda x: float(x))
df['ADR']    = df['ADR'].apply(lambda x: float(x))
df['KPR']    = df['KPR'].apply(lambda x: float(x))
df['APR']    = df['APR'].apply(lambda x: float(x))
df['FKPR']   = df['FKPR'].apply(lambda x: float(x))
df['FDPR']   = df['FDPR'].apply(lambda x: float(x))
df['K']      = df['K'].apply(lambda x: int(x))
df['D']      = df['D'].apply(lambda x: int(x))
df['A']      = df['A'].apply(lambda x: int(x))
df['FK']     = df['FK'].apply(lambda x: int(x))
df['FD']     = df['FD'].apply(lambda x: int(x))
df['Region'] = df['Region'].apply(lambda x: int(x))
df.head()
```

Saving to csv file...

```python
df.to_csv('vlr_player_data.csv', index=False)
```
