There appear to be some big outliers. This is likely due to the fact that some players may have only played an agent once or twice, but had insane games. Want to remove these to be more representative of the averages...


```python
df.shape[0]
```

```python
# filter out any data points with match number lower than 5

df = df[df['match_number'] >= 5]
```

```python
df.shape[0]
```

```python
plt.figure(figsize=(13,9))

sns.scatterplot(x=df['KD'], y=df['ADR'], hue=df['Agent'])
```

This seemed to get rid of a good amount of outliers. Saving the new data:


```python
df.to_csv('vlr_player_data_cleaned.csv', index=False)
```

Comparing Agents with each other

```python
df_cypher_jett = df[(df['Agent'] == 'cypher') | (df['Agent'] == 'jett')]

plt.figure(figsize=(13,9))
plt.title('Cypher and Jett ADR vs KD')

sns.scatterplot(x=df_cypher_jett['KD'], y=df_cypher_jett['ADR'], hue=df_cypher_jett['Agent'])
```

```python
plt.figure(figsize=(10,6))
plt.title('Cypher and Jett KD')
sns.kdeplot(data=df_cypher_jett, x='KD', hue='Agent', shade=True)
```

```python
plt.figure(figsize=(10,6))
plt.title('Cypher and Jett First Kills Per Round')
sns.kdeplot(data=df_cypher_jett, x='FKPR', hue='Agent', shade=True)
```

```python
plt.figure(figsize=(10,6))
plt.title('Cypher and Jett Rating')
sns.kdeplot(data=df_cypher_jett, x='Rating', hue='Agent', shade=True)
```

```python
# making new dataframes for each of the agent categories

df_sentinels = df[(df['Agent'] == 'cypher') | (df['Agent'] == 'killjoy') | (df['Agent'] == 'sage') | (df['Agent'] == 'chamber')]
df_duelists = df[(df['Agent'] == 'reyna') | (df['Agent'] == 'jett') | (df['Agent'] == 'raze') | (df['Agent'] == 'yoru') | (df['Agent'] == 'neon') | (df['Agent'] == 'phoenix')]
df_initiators = df[(df['Agent'] == 'skye') | (df['Agent'] == 'sova') | (df['Agent'] == 'breach') | (df['Agent'] == 'fade') | (df['Agent'] == 'kayo')]
df_controllers = df[(df['Agent'] == 'astra') | (df['Agent'] == 'viper') | (df['Agent'] == 'brimstone') | (df['Agent'] == 'harbor') | (df['Agent'] == 'omen')]
```

```python
plt.figure(figsize=(13,9))
plt.title('Sentinels ADR vs KD')

sns.scatterplot(x=df_sentinels['KD'], y=df_sentinels['ADR'], hue=df_sentinels['Agent'])
```

```python
plt.figure(figsize=(10,6))
plt.title('Sentinels First Kills Per Round')
sns.kdeplot(data=df_sentinels, x='FKPR', hue='Agent', shade=True)
```

```python
plt.figure(figsize=(10,6))
plt.title('Duelists First Kills Per Round')
sns.kdeplot(data=df_duelists, x='FKPR', hue='Agent', shade=True)
```

```python
plt.figure(figsize=(10,6))
plt.title('Initiators First Kills Per Round')
sns.kdeplot(data=df_initiators, x='FKPR', hue='Agent', shade=True)
```

```python
plt.figure(figsize=(10,6))
plt.title('Controllers First Kills Per Round')
sns.kdeplot(data=df_controllers, x='FKPR', hue='Agent', shade=True)
```
